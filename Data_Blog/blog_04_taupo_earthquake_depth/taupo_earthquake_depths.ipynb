```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.transforms as transforms
import pandas as pd
import geopandas as gpd
from shapely.geometry import Polygon, Point
from shapely import wkt

import requests
import json

import datetime as dt
import numpy as np
```

```python
xsize = 15
ysize = 5
maxdepth = '30'
search = '176.268494+-38.52276,176.270691+-39.123615,175.475281+-39.127024,175.479675+-38.512445,176.268494+-38.52276'
```

#### Data for 2022

```python
startdate = '2022-01-01T00:00:00.0Z'

now = dt.datetime.now()
start = dt.datetime.strptime(startdate, "%Y-%m-%dT%H:%M:%S.%fZ")
# titlestart = start.strftime("%Y-%m-%d %H:%M")
```

```python
url = "http://wfs.geonet.org.nz/geonet/ows?service=WFS&version=1.0.0&request=GetFeature&typeName=geonet:quake_search_v1&outputFormat=csv&cql_filter=origintime>=" + \
startdate + \
"+AND+WITHIN(origin_geom,POLYGON((" + search + "+)))+AND+depth<" + \
maxdepth

print (url)

cat = pd.read_csv(url, parse_dates=['origintime'])
```

```python
#eqs as a geopandas dataframe
eqs = gpd.GeoDataFrame(
    cat, geometry=gpd.points_from_xy(cat.longitude, cat.latitude), crs='EPSG:4326')
#compatible crs
eqs.to_crs(crs=2193, inplace=True)
```

```python
#dashed area used on VAB map
df = pd.read_csv('search_area_for_map.csv', names=['longitude','latitude'])

polygon_geom = Polygon(zip(df.longitude, df.latitude))
tsarea = gpd.GeoDataFrame(index=[0], crs='EPSG:4326', geometry=[polygon_geom])
tsarea.to_crs(crs=2193, inplace=True)
```

```python
#find only earthquakes within time-series area
#intersection
eqsts = gpd.overlay(eqs, tsarea, how='intersection')
```

```python
#remove fixed depths
depthdata = eqsts[eqsts['depthtype']!='operator assigned']
```

```python
print(len(eqsts))
print(len(depthdata))
print(100*len(depthdata)/len(eqsts))
```

```python
mean = depthdata['depth'].mean()
```

```python
median = depthdata['depth'].quantile(0.5)
```

```python
p5 = depthdata['depth'].quantile(0.05)
p95 = depthdata['depth'].quantile(0.95)
print (p5, p95)
```

```python
p25 = depthdata['depth'].quantile(0.25)
p75 = depthdata['depth'].quantile(0.75)
print (p25, p75)
```

```python
#data distribution only

fig,ax = plt.subplots(1,1, figsize=(5,7.5))
depthdata['depth'].hist(bins=list(range(21)), ax=ax, orientation='horizontal', color='cornflowerblue')
ax.invert_yaxis()
ax.set_xlabel('number of events')
ax.set_ylabel('depth (km)')
ax.set_yticks(list(range(0,22,2)))

fig.savefig('depth_distribution.png', bbox_inches='tight', facecolor='white')
```

```python
#with mean and median
fig,ax = plt.subplots(1,1, figsize=(5,7.5))
depthdata['depth'].hist(bins=list(range(21)), ax=ax, orientation='horizontal', color='cornflowerblue')
ax.invert_yaxis()
ax.set_xlabel('number of events')
ax.set_ylabel('depth (km)')
ax.set_yticks(list(range(0,22,2)))

ax.axhline(y=mean, color='#edbf64', linestyle='dashed', label='mean')
ax.axhline(y=median, color='#ed7b64', linestyle='dashed', label='median')

ax.legend(loc = 'lower right')

fig.savefig('depth_distribution_with_mean_median.png', bbox_inches='tight', facecolor='white')
```
