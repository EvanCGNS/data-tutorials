# Visualize Historic Volcanic Activity at Tongariro for Data Blog

Using data in https://github.com/GeoNet/data/tree/main/historic-volcanic-activity.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import squarify
```

```python
color = 'lightseagreen'
```

```python
url = 'https://raw.githubusercontent.com/GeoNet/data/main/historic-volcanic-activity/historic_eruptive_activity_tongariro.csv'
```

```python
df = pd.read_csv(url, index_col = 'Date LocalTime', parse_dates=True, usecols=['Date LocalTime', 'Location', 'Certainty'])
#keep only rows with a valid index
df = df.iloc[df.index.notnull()]
```

```python
df.head()
```

```python
first = df.index[0].strftime('%Y-%m-%d')
last = df.index[-1].strftime('%Y-%m-%d')
```

## Number of occurrences of each activity location

```python
location = pd.DataFrame(df['Location'].value_counts())
```

```python
#move the unknown index to the end, it looks better
location = location.drop('unknown').append(location.loc['unknown'])
```

```python
fig,ax = plt.subplots(1,1, figsize=(25,5))
ax.bar(location.index, location['Location'], color=color)
ax.set_ylabel('Number of occurrences')

plt.suptitle('Tongariro Historic Volcanic Activity - Activity Locations', fontsize=16)
plt.title(first+' - '+last)

#geonet attribution
ax.text(x = 0.9, y = 0.9, transform=fig.transFigure, s = 'Source:https://github.com/GeoNet/data', horizontalalignment='right')

fig.savefig('volcanic_activity_Tongariro-activity_locations_occurrences.png', facecolor='white')
```

## Proportion of occurrences of each activity location

```python
location.sort_values(by='Location', ascending=False, inplace=True)
location['group'] = ['Te Maari', 'Red Crater', 'Te Maari', 'Other', 'Other', 'Te Maari', 'Red Crater', 'Te Maari', 'Other', 'Other']
```

```python
locationgr = (location.groupby(by='group').sum()).sort_values(by='Location', ascending=False)
```

```python
locationgr.reset_index(inplace=True) 
```

```python
locationgr['label'] = locationgr['group'] + ' (' + locationgr['Location'].astype(str) + ')'
```

```python
locationgr
```

```python
colors=['lightseagreen', '#cfa219', '#cf1977']

fig,ax=plt.subplots(1,1, figsize=(12,6))
squarify.plot(ax=ax, sizes=locationgr['Location'], label=locationgr['label'], color=colors, alpha=0.6, pad=True)
ax.axis('off')

plt.suptitle('Tongariro Historic Volcanic Activity - Grouped Activity Locations', fontsize=16)
plt.title(first+' - '+last);

#geonet attribution
ax.text(x = 0.9, y = 0.9, transform=fig.transFigure, s = 'Source:https://github.com/GeoNet/data', horizontalalignment='right')

plt.savefig('volcanic_actvity_Tongariro_grouped_activity_locations.png', facecolor='white', bbox_inches='tight')
```

## Focus on three locations

```python
dfrc = df[df['Location']=='Red Crater']
dftm = df[df['Location']=='Te Maari']
dfutm = df[df['Location']=='Upper Te Maari']
```

```python
dfrc['occurrence'] = np.arange(len(dfrc))
dfrc['occurrence'] += 1

dftm['occurrence'] = np.arange(len(dftm))
dftm['occurrence'] += 1

dfutm['occurrence'] = np.arange(len(dfutm))
dfutm['occurrence'] += 1
```

```python
dfutm.head()
```

```python
#cumulative occurrence
fig,ax = plt.subplots(1,1, figsize=(15,5))

color2 = '#b26720'
color3 = '#b22025'

ax.step(dfrc.index, dfrc['occurrence'], where='post', color=color, label='Red Crater')
ax.step(dftm.index, dftm['occurrence'], where='post', color=color2, label='Te Maari')
ax.step(dfutm.index, dfutm['occurrence'], where='post', color=color3, label='Upper Te Maari')
ax.grid(alpha=0.3)
ax.set_ylabel('Number of activity reports from each location')

ax.legend(loc='upper left')

plt.suptitle('Tongariro Historic Volcanic Activity - Cumulative Reports of Activity at Most Active Locations', fontsize=16)
plt.title(first+' - '+last)

#geonet attribution
ax.text(x = 0.9, y = 0.9, transform=fig.transFigure, s = 'Source:https://github.com/GeoNet/data', horizontalalignment='right')

fig.savefig('volcanic_activity_Tongariro-cumulative_activity_most_active.png', facecolor='white', bbox_inches='tight')
```

## Activity at different locations over time

```python
#get list of locations, y-axis values for those, then map to a dataframe column
loclist = sorted(list(df['Location'].unique()))
yvals = range((len(loclist)),0,-1)
locvals = dict(zip(loclist, yvals))
#add column to dataframe based on dictionary key:value pairs
df['yval'] = df['Location'].map(locvals)
```

```python
fig,ax = plt.subplots(1,1, figsize=(15,5))

ax.plot(df['yval'], marker='^', markersize=5, linestyle='None', color=color)
    
#add locations as y-axis labels
ax.set_yticks(list(range((len(loclist)),0,-1)))
ax.set_yticklabels(loclist)

ax.grid(alpha=0.3)

plt.suptitle('Tongariro Historic Volcanic Activity - Occurrence of Activity at Different Locations', fontsize=16)
plt.title(first+' - '+last)

#geonet attribution
ax.text(x = 0.9, y = 0.9, transform=fig.transFigure, s = 'Source:https://github.com/GeoNet/data', horizontalalignment='right')

fig.savefig('volcanic_activity_Tongariro-activity_at_different_locations.png', facecolor='white', bbox_inches='tight')
```
